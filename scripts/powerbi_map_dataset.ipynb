```python
import pandas as pd
import numpy as np
```

```python
#Dataframe containing ips and exam results of high schools in france
ips_res_df = pd.read_csv("../data/ips_results.csv", delimiter=",", index_col=0)
ips_res_df.head()
```

```python
bins = np.arange(40, 160, 10)

map_dataset = ips_res_df.groupby('Département')['IPS Ensemble GT-PRO'].mean().reset_index()

#Creating bins so that power BI render can be used
for index, row in map_dataset.iterrows():
    for i in range(len(bins) - 1):
        if row["IPS Ensemble GT-PRO"] >= bins[i] and row["IPS Ensemble GT-PRO"] <= bins[i + 1]:
            map_dataset.loc[index, "IPS Ensemble GT-PRO"] = f"{bins[i]}-{bins[i + 1]}"

map_dataset.to_csv("../data/map_dataset.csv")
```
