```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
#Dataframe containing ips and exam results of high schools in france
ips_res_df = pd.read_csv("../data/ips_results.csv", delimiter=",", index_col=0)
ips_res_df.head()
```

```python
#Overlayed histogram of exam results per exam per year for public and private high schools
ips_res_df_per_year = ips_res_df.groupby(["Annee"])
#Making an individual graph for each year
#Use fig.show to demonstrate claims in next cell
for year in ips_res_df_per_year.groups:
    #If else to account for exam reforms that were applied in 2021
    if year < 2021:
        fig = px.histogram(ips_res_df_per_year.get_group(year), x="Taux de reussite", color="Secteur", nbins=40, barmode="overlay")
        fig.update_layout(xaxis_title="Taux de reussite", yaxis_title="Nombre de lycées",
                          legend_title="Secteur", title=f"Histogramme du taux de reussite pour toutes series pour {year} en fonction du secteur") 
        #fig.show()
    else:
        fig = px.histogram(ips_res_df_per_year.get_group(year), x="Taux de reussite", color="Secteur", nbins=40, barmode="overlay")
        fig.update_layout(xaxis_title="Taux de reussite", yaxis_title="Nombre de lycées",
                          legend_title="Secteur", title=f"Histogramme du taux de reussite pour la serie generale pour {year} en fonction du secteur") 
        #fig.show()
```

```python
#Most graphs have similar patterns with negligible values inferior to 70 (take off the three following lines to see negligible values)


fig = px.histogram(ips_res_df, x="Taux de reussite", color="Secteur", barmode="overlay", width=750)
fig.update_layout(xaxis_title="Taux de reussite", yaxis_title="Nombre de lycées",
                  legend_title="Secteur", title=f"Taux de reussite en fonction du secteur") 
fig.show()
```
