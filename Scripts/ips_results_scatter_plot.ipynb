```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
#Dataframe containing ips and exam results of high schools in france
ips_res_df = pd.read_csv("ips_results.csv", delimiter=",", index_col=0)
ips_res_df.head()
```

```python
plt.scatter(ips_res_df['Taux de reussite - Gnle'], ips_res_df['IPS Ensemble GT-PRO'], s = 0.2)
plt.title("correlation entre l'IPS et le taux de reussite au BAC")
plt.xlabel('Taux de reussite - Gnle')
plt.ylabel('IPS Ensemble GT-PRO')


plt.show()
```
